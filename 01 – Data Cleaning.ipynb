```python
df["TotalCharges"] = pd.to_numeric(

    df["TotalCharges"],

    errors="coerce"

)
```

```python
df.isnull().sum()
```

```python
df[df["TotalCharges"].isnull()]
```

```python
df = df.dropna()
```

```python
df.reset_index(

    drop=True,

    inplace=True

)
```

```python
print(

    "Duplicate Rows :",

    df.duplicated().sum()

)
```

```python
df["Churn"] = df["Churn"].map({

    "Yes":1,

    "No":0

})
```

```python
df.dtypes
```

```python
df["TenureGroup"] = pd.cut(

    df["tenure"],

    bins=[

        0,

        12,

        24,

        48,

        60,

        72

    ],

    labels=[

        "0-12 Months",

        "13-24 Months",

        "25-48 Months",

        "49-60 Months",

        "61-72 Months"

    ]

)
```

```python
df["MonthlyChargeCategory"] = pd.qcut(

    df["MonthlyCharges"],

    q=3,

    labels=[

        "Low",

        "Medium",

        "High"

    ]

)
```
